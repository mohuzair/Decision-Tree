```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
iris = load_iris()
irisData = iris.data
colNam = iris.feature_names
df = pd.DataFrame(irisData,columns=colNam)
df.head(10)
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
fig,axes=plt.subplots(2,2,figsize=(10,10))
df['sepal length (cm)'].hist(ax=axes[0][0])
df['sepal width (cm)'].hist(ax=axes[0][1])
df['petal length (cm)'].hist(ax=axes[1][0])
df['petal width (cm)'].hist(ax=axes[1][1])
plt.show()
```

```python
df.corr()
```

```python
X = df
Y = iris.target
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.4,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
print("Accuracy:",(accuracy_score(Y_test, Y_pred))*100 , "%")
```

```python
plt.figure(figsize=(20,20))
tree.plot_tree(model,fontsize=11)
```
